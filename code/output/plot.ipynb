```python
import json
import matplotlib.pyplot as plt
```

```python
experts = ['lru', 'lfu']
num_samples = 5
reward_type = 'lecar'
fname = 'adasample-{}-s{}-r{}.json'.format(experts, num_samples, reward_type)
with open(fname, 'r') as f:
    w_hist = json.load(f)

w_lru = [i[0] for i in w_hist]
w_lfu = [i[1] for i in w_hist]
# w_hyper = [i[2] for i in w_hist]
plt.plot(w_lru)
plt.plot(w_lfu)
# plt.plot(w_hyper)
plt.show()
```

```python
experts = ['lru', 'lfu', 'hyperbolic']
num_samples = 20
reward_type = 'lecar'
fname = 'adasample-{}-s{}-r{}.json'.format(experts, num_samples, reward_type)
with open(fname, 'r') as f:
    w_hist = json.load(f)

w_lru = [i[0] for i in w_hist]
w_lfu = [i[1] for i in w_hist]
w_hyper = [i[2] for i in w_hist]
plt.plot(w_lru)
plt.plot(w_lfu)
plt.plot(w_hyper)
plt.show()
```

```python
experts = ['lru', 'lfu']
num_clients = 128
num_samples = 20
reward_type = "lecar"
fname = 'paraadasample-{}-s{}-r{}-c{}.json'.format(experts, num_samples, reward_type, num_clients)
with open(fname, 'r') as f:
    w_hist = json.load(f)

w_lru = [i[0] for i in w_hist]
w_lfu = [i[1] for i in w_hist]
fig = plt.figure()
fig.patch.set_facecolor('white')
ax = fig.add_subplot(111)
ax.plot(w_lru)
ax.plot(w_lfu)
plt.show()
```
